```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import SimpleITK as sitk

def imread(fpath):
    if fpath.endswith('.list'):
        with open(fpath,'r') as f:
            dicom_names = [x for x in f.read().split('\n') if len(x) > 0]
        if not os.path.exists(dicom_names[0]):
            dicom_names = [os.path.join(os.path.dirname(fpath),x) for x in dicom_names]
        reader = sitk.ImageSeriesReader()
        reader.SetFileNames(dicom_names)
    else:
        reader= sitk.ImageFileReader()
        reader.SetFileName(fpath)
    img = reader.Execute()
    arr = sitk.GetArrayFromImage(img)    
    spacing = img.GetSpacing()
    origin = img.GetOrigin()
    direction = img.GetDirection()    
    return arr,spacing,origin,direction

def imwrite(fpath,arr,spacing,origin,direction,use_compression=True):
    img = sitk.GetImageFromArray(arr)
    img.SetSpacing(spacing)
    img.SetOrigin(origin)
    img.SetDirection(direction)
    writer = sitk.ImageFileWriter()    
    writer.SetFileName(fpath)
    writer.SetUseCompression(use_compression)
    writer.Execute(img)
```

```python
from scipy import ndimage
```

```python
lll = mask==1
print(np.sum(lll))
lll = ndimage.binary_dilation(lll,iterations=1)
lll = lll.astype(int)
print(np.sum(lll))

lul = mask==2
print(np.sum(lul))
lul = ndimage.binary_dilation(lul,iterations=1)
lul = lul.astype(int)
print(np.sum(lul))
```

```python
assumed_fissure_location =  ((lll+lul )==2)
print(np.sum(assumed_fissure_location))
```

```python
tmp = np.sum(assumed_fissure_location,axis=0)
plt.imshow(tmp,cmap='gray')
```

```python
dimg = img.astype(np.float)
dimg[assumed_fissure_location==0]=np.nan
```

```python
for n in range(3):
    plt.figure(n)
    tmp = np.nansum(dimg,axis=n)
    plt.imshow(tmp,cmap='gray')
    plt.title(f"projected axis {n}")
```
